```python
# Import necessary libraries
import folium
import leafmap.foliumap as leafmap_folium

# Initialize the map at the desired center and zoom level
m = leafmap_folium.Map(center=[40, -100], zoom=4)

# URLs for the WMS layers
url1 = "https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2021_Impervious_L48/wms?"
url2 = "https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2001_Impervious_L48/wms?"

# WMS Tile Layer for NLCD 2021
left_layer = folium.WmsTileLayer(
    url=url1,
    layers="NLCD_2021_Impervious_L48",
    name="NLCD 2021",
    attr="MRLC",
    fmt="image/png",
    transparent=True,
)

# WMS Tile Layer for NLCD 2001
right_layer = folium.WmsTileLayer(
    url=url2,
    layers="NLCD_2001_Impervious_L48",
    name="NLCD 2001",
    attr="MRLC",
    fmt="image/png",
    transparent=True,
)

# Add the layers to the map
left_layer.add_to(m)
right_layer.add_to(m)



# split_map function to compare layers
m.split_map(left_layer, right_layer)

# Display the map
m
```
